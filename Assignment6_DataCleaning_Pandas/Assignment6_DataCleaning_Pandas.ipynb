Some values in the the FlightNumber column are missing. These numbers are meant to increase by 10 with each row so 10055 and 10075 need to be put in place. Fill in these missing numbers and make the column an integer column (instead of a float column).

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
                               'Budapest_PaRis', 'Brussels_londOn'], 
                   'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                   'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', ' (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})
df
```

### Question 5:
In the RecentDelays column, the values have been entered into the DataFrame as a list. We would like each first value in its own column, each
second value in its own column, and so on. If there isn't an Nth value, the value should be NaN. Expand the Series of lists into a DataFrame named delays, rename the columns delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df with delays.

```python
#just taken again the given data frame in the question
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
                               'Budapest_PaRis', 'Brussels_londOn'], 
                   'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                   'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', ' (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})
df
```

```python
# Reading RecentDelays column and converting the list format into data frame
df_new = pd.DataFrame(df['RecentDelays'].values.tolist())
df_new
```

```python
# One way to rename column names in dataframe
df_new1= df_new.copy()
df_new1.rename( columns={0:'Delay_1',1:'Delay_2',2:'Delay_3'}, inplace=True)
df_new1
```

```python
# 2nd way of renaming 
# By using shape of df_new dataframe, creating column name dynamically like Delay_1,Delay-2 ....and so on using for loop and 
#storing in the dictonary format.

col_dict ={}
for i in range(df_new.shape[1]):
    Key = df_new.columns[i]
    #print(key,i)
    Value = "Delay_" + str(i+1)
    col_dict[Key] = Value

    
col_dict
```

```python
# Renaming df_new column names to the values in col_dict
df_new.rename(columns=col_dict,inplace=True)
df_new
```

```python
#Adding the new columns to the old data frame
df[["Delay_1","Delay_2","Delay_3"]] = df_new[["Delay_1","Delay_2","Delay_3"]]
```

```python
df
```

```python
#Dropping the RecentDelays column from the data frame
df.drop('RecentDelays',axis=1,inplace=True)
df
```
